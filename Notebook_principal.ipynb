### Acciones a tomar

Se tomara la opcion de eliminar los registros que no se identificaron y ademas los campos que no aportan nada al resultado que se planea obetenr. Ademas de separa el campo "geo_point" en dos campos denominados: _latitude_ y _longitude_. Por ultimo se renombrara las columnas.


Realizacion de operaciones:

```python
# Eliminamos ultimo registro
data_1 = data.drop([733],axis=0).copy()

# Mostrar resultado
data_1.tail()
```

```python
# Eliminacion de las campo que no aportan informacion
data_2 = data_1.drop(['nueva_clave', 'cuartil', 'quintil','octil', 'decil', 'geo_shape'],axis=1).copy()

#Mostrar resutados
data_2.head()
```

```python
# Limpieza del campo "geo_point"

data_2['geo_point']=data_2['geo_point'].str.replace('"type": "Point", "coordinates":','')
data_2['geo_point']=data_2['geo_point'].str.replace('[','') 
data_2['geo_point']=data_2['geo_point'].str.replace('{','') 
data_2['geo_point']=data_2['geo_point'].str.replace(']','') 
data_2['geo_point']=data_2['geo_point'].str.replace('}','') 
```

```python
# Separacion de latitud y longitud
data_2[['longitude','latitude']] = data_2['geo_point'].str.split(",",expand=True)

# Eliminacion de columna "geo_point_2d"
data_3 = data_2.drop(['geo_point'],axis=1).copy()

# Mostrar resultados 
data_3.head()
```

```python
# Renombramiento de columnas

#Diccionario con nuevo nombres
diccionario_cabeceras = {
    'alcandia':'alcaldia',
    'colonia':'colonia',
    'total':'casos_totales',
    'latitude':'latitude',
    'longitude':'longitude'
}

#Aplicaion del diccionario
data_4 = data_3.rename(columns=diccionario_cabeceras)

# Mostrar resultados 
data_4.head()
```

### Acciones a tomar

Se eliminaran los campos id, geo_shape, name, descriptio, lon, lat y linea_esta. Ademas de separa el campo "geo_point_2d" en dos campos denominados: _latitude_ y _longitude_. Por ultimo se renombrara las columnas.


Realizacion de operaciones:

```python
# Eliminacion de las campo que no aportan informacion
data_metro_1 = data_metro.drop(['id', 'geo_shape', 'lon', 'lat', 'linea_esta'],axis=1).copy()

#Mostrar resutados
data_metro_1.head()
```

```python
# Separacion de latitud y longitud
data_metro_1[['latitude','longitude']] = data_metro_1['geo_point_2d'].str.split(",",expand=True)

# Eliminacion de columna "geo_point_2d"
data_metro_2 = data_metro_1.drop(['geo_point_2d'],axis=1).copy()

# Mostrar resultados 
data_metro_2.head()
```

```python
# Renombramiento de columnas

#Diccionario con nuevo nombres
diccionario_cabeceras_metro = {
    'name':'nombre_estacion',
    'descriptio':'estacion',
    'latitude':'latitude',
    'longitude':'longitude'
}

#Aplicaion del diccionario
data_metro_3 = data_metro_2.rename(columns=diccionario_cabeceras_metro)

# Mostrar resultados 
data_metro_3.head()
```
